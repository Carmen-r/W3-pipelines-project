Realizaré un scraping de la página web `IMDb` para sacar el rating de las películas con mejor valoración

### Librería

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import scr.limpieza as lt
```

```python
url = "https://www.imdb.com/chart/moviemeter/?sort=ir,desc&mode=simple&page=1"
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
titles = soup.find_all("td", {"class": 'titleColumn'})
```

```python
rating = soup.find_all("td", {"class": 'ratingColumn imdbRating'})
```

```python
#Es importante que todas las listas tengan el mismo número de filas 
lista_titles = [a.getText().strip() for a in titles]
lista_t = lista_titles
len(lista_t)
```

```python
lista_rating = [precio.getText().strip() for precio in rating]
lista_r = lista_rating
len(lista_r)
```

```python
#Creamos un diccionario para hacer un DataFrame
diccionario = {"Title": lista_t, "Rating": lista_r}
```

```python
df1 = pd.DataFrame(diccionario)
df1.head()
```

```python
#Con esta función nos quedamos con los títulos de las peliculas
def limpia_precio(x):
    x = x.split("\n")
    x = x[0].split(" .")
    return x[0] + " ."

df1.Title = df1.Title.apply(limpia_precio)
df1
```

```python
list_ = ['Joker .', 'Thor: Ragnarok .', 'Vengadores: Infinity War .', 'El juicio de los 7 de Chicago .', 'Black Panther .']
```

```python
#Con esta función nos quedamos con aquellas películas de IMDb que coincid
df2 = lt.create(df1,'Title',list_)
df2
```
