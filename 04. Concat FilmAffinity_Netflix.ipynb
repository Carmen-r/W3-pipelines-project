### 1. Netflix

```python
data = pd.read_csv("midatasetlimpio/netflix_clean.csv", encoding = "ISO-8859-1")
data
```

```python
#Me quedo con aquellas columnas que me interesan
columnas = ['title', 'director']
df1 = data[columnas]
df1
```

```python
#Buscamos los directores de FilmAffinity que están en el dataset de Netflix
list_regex = [r'^francis ford coppola$', r'^ernest r. dickerson$', r'^vince gilligan$', r'^steven spielberg$', r'^james l. brooks$']
directores = ['francis ford coppola', 'ernest r. dickerson', 'vince gilligan', 'steven spielberg', 'james l. brooks']
```

```python
df1['Directores'] = df1.director.str.lower().replace(list_regex, directores, regex=True)
```

```python
df1
```

```python
list_ = ['francis ford coppola', 'ernest r. dickerson', 'vince gilligan', 'steven spielberg', 'james l. brooks']
```

```python
#Limpiamos la columna para quedarnos con los directores que interesan
df = lt.create(df1,'Directores',list_)
df.Directores.value_counts()
```

```python
df_ = df.drop(['Directores'], axis=1)
df_
```

```python
#Se renombra el nombre de las columnas para poder hacer la concatenación
diccio_columnas = {"title": "Peliculas", "director": "Director"}
df_.rename(columns=diccio_columnas, inplace = True)
data = df_.rename(columns = diccio_columnas)
data
```
